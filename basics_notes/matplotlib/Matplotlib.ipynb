# Matplotlib
    * Matplotlib is the most popular plotting library for Python.
    * It gives you control over every aspect of a figure.
    * It was designed to have a similar feel to MatLab's graphical plotting.

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
x
```

```python
y = x ** 2
y
```

```python
# Functional Method To Create Plot
plt.plot(x, y)
```

```python
# Can Use MatLab's Style Arguments
plt.plot(x, y, 'r-')
```

```python
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
# Can Create Multiple Plots on Same Canvas Using Subplot Function.
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
# Object Oriented Way To Create Plot - Mostly Used (RECOMMENDED)
fig = plt.figure()

# fig.add_axes([left, bottom, width, height of axes])
axes = fig.add_axes([0.5, 0.5, 0.8, 0.8])
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes_one = fig.add_axes([0.1, 0.1, 0.8, 0.8])
# 20% - From Left
# 50% - From Bottom
# 40% - Of Parent Axes Width
# 30% - Of Parent Axes Height----
axes_two = fig.add_axes([0.2, 0.5, 0.4, 0.3])
```

```python
fig = plt.figure()

axes_one = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes_two = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes_one.plot(x, y)
axes_one.set_title('Larger Plot')

axes_two.plot(y, x)
axes_two.set_title('Smaller Plot')
```

```python
# Creating Subplots Using OOP Way
fig, axes = plt.subplots(nrows=1, ncols=2)

# axes.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3)

# Fix Overlapping
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for current_ax in axes:
    current_ax.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y)
axes[0].set_title('First Plot')
axes[0].set_xlabel('First X')
axes[0].set_ylabel('First Y')
axes[1].plot(y, x)
axes[1].set_title('Second Plot')
axes[1].set_xlabel('Second X')
axes[1].set_ylabel('Second Y')

fig.tight_layout()
```

## Figure Size, Aspect Ratio and DPI

```python
# figsize - in Inches - (width, height)
# dpi - pixel / inch
fig = plt.figure(figsize=(5, 3), dpi=100)
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(6, 3))
axes[0].plot(x, y)
axes[0].set_title('Major Plot')

axes[1].plot(y, x)
axes[1].set_title('Minor Plot')

fig.tight_layout()
```

## Add a Legend
    * Need Label on Plot

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, x ** 2, label='X Squared')
axes.plot(x, x ** 3, label='X Cubed')

axes.legend(loc=0) # Check Docs For More location options.
# Can Also Use Tuple As loc argument as a location for legend. loc = (left, bottom)
```

# Part - 3

## Plot Appearance

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])

# Color --> Can Use Some General Color Strings and Can Use Hex Colors.
# linewidth --> Control Line Width
# alpha --> Control Opacity
# linestyle --> Control Style of Line ('--', ':', '-', etc.)
# drawstyle --> ('steps', etc.)
# marker --> Show Marker of DataPoints ('o', '*', '+', '1')
# markersize --> Control Size of Marker
# markerfacecolor --> Control Marker Main Color
# markeredgecolor --> Control color of Marker's Face
# markeredgewidth --> Control Width of Marker Edge

# axes.plot(x, y, color='orange')
axes.plot(
    x, y, color="#EC8ADF", linewidth=3, 
    alpha=1, linestyle='-', drawstyle='steps', 
    marker='o', markersize=20,
    markerfacecolor='yellow', markeredgewidth=3,
    markeredgecolor='green'
)
```

## Control Over Axis Appearance

### Plot Range

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y, color='purple', lw=2, ls='--')

axes.set_xlim([0, 1])
axes.set_ylim([0, 2])
```
